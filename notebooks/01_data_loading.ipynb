```python
import pandas as pd
import numpy as np
import os
from IPython.display import display

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 1000)

print("Libraries imported successfully.")
```

```python
RAW_DATA_PATH = "../data/raw/"
PROCESSED_DATA_PATH = "../data/processed/"

# Create processed folder if it doesn't exist
os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)

print("Paths defined successfully.")
print("Raw Data Path:", RAW_DATA_PATH)
print("Processed Data Path:", PROCESSED_DATA_PATH)
```

```python
files = os.listdir(RAW_DATA_PATH)

print("Files found in raw data folder:")
for file in files:
    print(file)
```

```python
city_df = pd.read_excel(RAW_DATA_PATH + "City.xlsx")
continent_df = pd.read_excel(RAW_DATA_PATH + "Continent.xlsx")
country_df = pd.read_excel(RAW_DATA_PATH + "Country.xlsx")
item_df = pd.read_excel(RAW_DATA_PATH + "Item.xlsx")
mode_df = pd.read_excel(RAW_DATA_PATH + "Mode.xlsx")
region_df = pd.read_excel(RAW_DATA_PATH + "Region.xlsx")
transaction_df = pd.read_excel(RAW_DATA_PATH + "Transaction.xlsx")
type_df = pd.read_excel(RAW_DATA_PATH + "Type.xlsx")
updated_item_df = pd.read_excel(RAW_DATA_PATH + "Updated_Item.xlsx")
user_df = pd.read_excel(RAW_DATA_PATH + "User.xlsx")

print("All Excel files loaded successfully.")
```

```python
item_df = updated_item_df.copy()

print("Using Updated_Item.xlsx as the final item dataset.")
print("Item dataset shape:", item_df.shape)
```

```python
datasets = {
    "city": city_df,
    "continent": continent_df,
    "country": country_df,
    "item": item_df,
    "mode": mode_df,
    "region": region_df,
    "transaction": transaction_df,
    "type": type_df,
    "user": user_df
}

print("Datasets dictionary created successfully.")
print("Total datasets:", len(datasets))
```
